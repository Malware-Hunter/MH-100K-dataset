```python
def load_npz(path_npz='mh100.npz'):
    data = np.load(path_npz, allow_pickle=True)
    metadata = dataframe(data['metadata'], columns=data['metadata_columns'])
    
    columns_names = data['column_names']
    sha256 = data['sha256']

    # labels_ohe = OneHotEncoder().fit_transform(np.expand_dims(metadata['CLASS'].values, axis=1)).toarray()
    df_data = dataframe(data['data'], columns=columns_names)
    print(data['data'].shape, metadata.shape)

    labels = metadata['CLASS'].values.astype(np.int8)
    
    print(metadata['CLASS'].value_counts())
    return df_data, labels, metadata
```

## CHI-Square

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# def get_chi2(X, y):
#     return chi2_stats, p_values
```

```python
chi2_stats, p_values =  chi2(data.values, labels)  # Virus total scanners detections >= 4
```

```python
chi2_stats.shape, p_values.shape
```

```python
data.columns
```

```python
df_chi2 = dataframe({
    'names': data.columns,          # from load_mh100()
    'stats': chi2_stats,
    'p_values': p_values
})
```

```python
df_chi2.head(10)
```

```python
chi2_sorted = df_chi2.sort_values(by='stats', ascending=False).dropna()
chi2_sorted
```

```python
chi2_features = chi2_sorted[chi2_sorted['p_values'] < 0.05]  ## significance level (e.g. α = .05)
chi2_features
```
